# Unblinded entorno

```python
from funciones_navegacion import *
import os
import google.generativeai as genai
from dotenv import load_dotenv
from PIL import Image
```

```python
#  Inicializar el cliente con API Key
load_dotenv()
API_KEY = os.getenv("GEMINI_API_KEY")
genai.configure(api_key=API_KEY)
model = genai.GenerativeModel('models/gemini-2.5-flash')
```

```python
def ask_gemini(prompt = None, img = None):

    try:
        if img is None:
            print('Mode 1')
            response = model.generate_content(prompt)
            return response.text
        elif prompt is None:
            print('Mode 2')
            base_prompt = "Soy una persona ciega que quiero que me describas lo mas importante de esta imagen:"
            sec_prompt = "Si encuentras algún peligro en la imagen indicalo de forma concisa, en caso contrario describe la imagen lo mejor que puedas"
            response = model.generate_content([base_prompt, img, sec_prompt])
            return response.text
        else:
            print('Mode 3')
            base_prompt = "Soy una persona ciega que quiero obtener información de la siguiente imagen:"
            response = model.generate_content([base_prompt, img, prompt])
            return response.text
    except Exception as e:
        return f"Error en la petición: {e}"
```

```python
# Modo 1: Obtencion de una respuesta a un prompt
pregunta = "Explícame brevemente qué es una API como si tuviera 5 años."
resultado = ask_gemini(prompt = pregunta)

print("🤖 Gemini dice:")
print("-" * 20)
print(resultado)
```

```python
# Modo 2: Obtencion de una respuesta con una imagen
img = Image.open('praderas.jpg')
resultado = ask_gemini(img=img)

print("🤖 Gemini dice:")
print("-" * 20)
print(resultado)
```

```python
# Modo 3: Obtencion de una respuesta especifica con un prompt y una imagen
pregunta = "Dime que tipos de animales se ven en la imagen."
img = Image.open('praderas.jpg')
resultado = ask_gemini(prompt = pregunta, img = img)

print("🤖 Gemini dice:")
print("-" * 20)
print(resultado)
```
